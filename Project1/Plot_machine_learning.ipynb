```python
def plot_s_and_b(dataframe, feature1, feature2):
    markers = {'s':'+', 'b':'o'}
    s = dataframe.loc[dataframe['Prediction'].str.contains('s')]
    b = dataframe.loc[dataframe['Prediction'].str.contains('b')]
    sns.regplot(x=s[feature1], y=s[feature2], marker = 'o')
    sns.regplot(x=b[feature1], y=b[feature2], marker = '^')
```
